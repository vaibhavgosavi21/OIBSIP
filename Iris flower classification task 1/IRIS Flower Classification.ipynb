                                            COORELATION MATRIX 
A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. The value is in the range of -1 to 1. If two variables have high correlation, we can neglect one variable from those two.

```python
numeric_columns=df.select_dtypes(include='number')#select only numeric column
correlation_matrix=numeric_columns.corr() #calculate correaltion matrix
print(correlation_matrix)
```

```python
import seaborn as sns
numeric_columns=df.select_dtypes(include=['float64','int64'])
corr = numeric_columns.corr()
fig, ax = plt.subplots(figsize=(5,4))
sns.heatmap(corr, annot=True, ax=ax, cmap='coolwarm')
plt.show()
```

                                        LABEL ENCODER
In machine learning, we usually deal with datset which contain multiple labels in one or more than one columns. These labels can be in the form of words or numbers. Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import pandas as pd
```

```python
# Use LabelEncoder to encode the 'Species' column
df['Species'] = le.fit_transform(df['Species'])

# Display the  DataFrame
print(df.head())
```

                                        MODEL TRAINING 

```python
from sklearn.model_selection import train_test_split
#train - 70
#test - 30
x= df.drop(columns=['Species'])
y= df['Species']
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.30)
           
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
 #print metric to get performance
print("Accuracy:",model.score(x_test,y_test)*100)
```

```python
#knn- k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
#print metric to get performance
print("Accuracy:",model.score(x_test,y_test)*100)
```

```python
 #decision tree
 from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
#print metric to get performance
print("Accuracy:",model.score(x_test,y_test)*100)
```
